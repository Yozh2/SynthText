# Performance evaluation of the FCRN Depth Estimator


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
!ls images -l
```

### Load original image and predicted depth (as image and numpy.array)


```python
image_cv2 = cv2.imread('images/curved_paper.jpg')
depth_cv2 = cv2.imread('images/curved_paper.jpg_depth.png')
depth_npy = np.load('images/curved_paper.jpg_depth.npy').squeeze() # remove singled dimensions
```

### Let's look at the original image


```python
print(image_cv2.shape)
image_plt = image_cv2[:,:,::-1] # BGR -> RGB
fig, ax = plt.subplots()
im = ax.imshow(image_plt)
plt.show()
```

### Let's check how numpy depth is represented (should look simillar to depth_as_image)

```python
print(depth_npy.shape)
fig, ax = plt.subplots()
im = ax.imshow(depth_npy)
plt.show()
```

### Resize the depth image up to the original one

```python
dims = image_cv2.shape[1], image_cv2.shape[0]
depth_npy_resized = np.expand_dims(depth_npy, axis=2)
depth_cv2_resized = cv2.resize(depth_npy_resized, dims, interpolation=cv2.INTER_AREA)

print(depth_cv2_resized.shape)
fig, ax = plt.subplots()
im = ax.imshow(depth_cv2_resized)
plt.show()
```
